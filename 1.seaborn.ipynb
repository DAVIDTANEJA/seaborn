##  Joint Kernel

###### Bivariate Data -- is a data on each of two variables , where each value of one var. is paired with a value of other var.
###### on a scatter plot could be used to study the relationship b/w stride length and length of legs.

```python
import numpy as np

sns.set(style="white")

# Generate a random correlated bivariate dataset
random_state = np.random.RandomState(10)

mean = [2, 2]

cov = [(1, .5), (.5, 1)]

x1, x2 = random_state.multivariate_normal(mean, cov, 500).T

x1 = pd.Series(x1, name="X_1")
x2 = pd.Series(x2, name="X_2")

# Show the joint distribution using kernel density estimation
example3 = sns.jointplot(x1, x2, kind="kde", height=7, space=0, color='b')         # kde- kernel density
```

### Sinewaves

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def s(flip=2):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip)
        
sns.set()
s()
plt.show()
```

#### Sinewave - using generator

```python
def s(flip=2):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        yield(plt.plot(x, np.sin(x + i * 0.5) * (7 - i) * flip))         # yield
        
sns.set()
s = s()
plt.show()

print(next(s))   # print 1 wave
print(next(s))   # prints 2nd wave
```
